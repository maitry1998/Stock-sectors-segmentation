```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
# Load the data from the uploaded CSV file
file_path = '/Users/maitrychauhan/Desktop/stockdata/stocks.csv'
stocks_data = pd.read_csv(file_path)

# # Display the first few rows of the dataset to understand its structure
stocks_data.head(), stocks_data.info()

# Let's try plotting the Distribution of Market Capitalization again
plt.figure(figsize=(10, 6))
sns.histplot(stocks_data['market_cap'], bins=30, kde=True, color='blue')
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (in trillions)')
plt.ylabel('Frequency')
plt.xscale('log')  # Using logarithmic scale for better visualization
plt.xticks([1e9, 1e10, 1e11, 1e12, 1e13], ['1B', '10B', '100B', '1T', '10T'])
plt.grid(True)
plt.show()

```

```python
# Top 10 companies by Market Cap
top_companies = stocks_data.nlargest(10, 'market_cap')

plt.figure(figsize=(12, 6))
sns.barplot(x='market_cap', y='company_name', data=top_companies, palette='viridis')
plt.title('Top 10 Companies by Market Capitalization')
plt.xlabel('Market Capitalization (in trillions)')
plt.ylabel('Company Name')
plt.xscale('log')
plt.xticks([1e11, 1e12, 1e13], ['100B', '1T', '10T'])
plt.grid(True)
plt.show()
```

```python
# Distribution of Stock Price Changes
plt.figure(figsize=(10, 6))
sns.histplot(stocks_data['change'], bins=30, kde=True, color='red')
plt.title('Distribution of Stock Price Changes')
plt.xlabel('Change in Stock Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Scatter plot of Volume vs. Market Cap
plt.figure(figsize=(10, 6))
sns.scatterplot(x='market_cap', y='volume', data=stocks_data, color='green', alpha=0.6)
plt.title('Trade Volume vs. Market Capitalization')
plt.xlabel('Market Capitalization (in trillions)')
plt.ylabel('Trade Volume')
plt.xscale('log')
plt.yscale('log')  # Using log scale for better visibility on large ranges
plt.xticks([1e9, 1e10, 1e11, 1e12, 1e13], ['1B', '10B', '100B', '1T', '10T'])
plt.grid(True)
plt.show()
```

The scatter plot above visualizes the relationship between trade volume and market capitalization, both scales are logarithmic to handle the wide range of values effectively. There doesn't seem to be a strong linear correlation as data points are spread across the plot, indicating varying trade volumes for companies of similar sizes.

```python
# Clean the pe_ratio_ttm column and convert it to float
stocks_data['pe_ratio_ttm'] = pd.to_numeric(stocks_data['pe_ratio_ttm'], errors='coerce')

# Distribution of PE Ratio
plt.figure(figsize=(10, 6))
sns.histplot(stocks_data['pe_ratio_ttm'].dropna(), bins=30, kde=True, color='purple')
plt.title('Distribution of PE Ratios (TTM)')
plt.xlabel('PE Ratio')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Most companies have PE ratios clustering in the lower range, with a long tail indicating a few companies with significantly higher ratios. This spread suggests a variation in how investors value earnings across different companies.
